```
#criando uma coluna para deixar todas as strings em maiúsculo
cidades['Descricao'] = cidades['nome'].str.upper()
```

```
#usando uma função para remover acentos de strings - fonte https://wiki.python.org.br/RemovedorDeAcentos
from unicodedata import normalize

def remover_acentos(txt):
  return normalize('NFKD', txt).encode('ASCII', 'ignore').decode('ASCII')
if __name__ == '__main__':
  from doctest import testmod
  testmod()
```

```
#criando uma lista com os nomes alterados
lista = []
for value in cidades['Descricao'].values:
  lista.append(remover_acentos(value))
```

```
#criando uma nova coluna com a lista criada
cidades['descricao'] = lista
```

```
#conferindo a transformação
cidades.head()
```

```
#criando um subset só com as colunas que eu preciso
cidade_estado = cidades[['descricao', 'codigo_uf']]
cidade_estado.columns = ['Descricao', 'UF']
```

### Desafio: Estruturar uma função para a limpeza da tabela e a junção da tabela população e gastos e populaçao recente

```
def manipula_dados(dados_como_string, coluna_int, coluna_index ,nome_novo_int, nome_novo_index, dataset_join):
  """Lê arquivo de origem string para criação de um dataframe com novos dados, 
     Transforma uma coluna de string para números inteiros, excluindo os espaços e retira os espaços de uma coluna de strings. 
     Cria um dataframe apenas com as duas colunas e transforma a coluna de strings em índice
     Trata índice com nome duplicado
     Ajusta o índice do dataset_join para que conste apenas a partir da 3 letra
     Unifica o dataframe criado com o dataframe desejado.

   Variáveis: 
    dados_como_string -> String com os dados do dataframe desejado
    coluna_int -> string, nome da coluna a transformar para números inteiros
    coluna_index -> string, nome da coluna que será transformada em índice
    nome_novo_ind -> string, novo nome da coluna transformada em inteiro
    nome_novo_índice -> String, novo nome da coluna que será transformada em índice
    dataset_join -> pd.Dataframe, dataframe com os dados a serem unificados ao dataset transformado 

   Requisitos:
    import pandas as pd
    from io import StringIO

   Retorno:
     Uma tupla com dois datasets, o transformado e o unido
  """
  #importando as bibliotecas necessárias
  import pandas as pd
  from io import StringIO
  #transforma o arquivo em algo legível para o read_csv
  origem = StringIO(dados_como_string)
  #criando o dataframe
  dataframe_origem = pd.read_csv(origem,sep='\t')
  #excluindo os nulos
  dataframe_destino = dataframe_origem.dropna()
  #criando a coluna para número inteiro
  dataframe_destino[nome_novo_int] = dataframe_destino[coluna_int].str.replace(" ", "").astype(int)
  #criando a coluna de índice já excluindo os espaços antes e depois dos nomes
  dataframe_destino[nome_novo_index] = dataframe_destino[coluna_index].str.strip()
  #aproveitando apenas as colunas criadas e ajustando o índice
  dataframe_final = dataframe_destino[[nome_novo_index, nome_novo_int]].set_index(nome_novo_index)
  #alterando o índice do dataset a ser unido
  dataset_join.index = dataset_join.index.str[3:]
  #ajustando o caso dos nomes repetidos
  for indice in dataset_join.index:
    dataframe_final.index = dataframe_final.index.str.replace(f'{indice} {indice}', indice)
  #unindo os datasets
  unido = dataframe_final.join(dataset_join)
  return (dataframe_final, unido) 
```
